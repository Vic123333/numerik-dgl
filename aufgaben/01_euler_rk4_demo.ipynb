# Numerische DGL-Lösung: Euler & Runge-Kutta 4

**Aufgabe:** Löse das AWP
$$y' = -y, \quad y(0) = 1$$
mit exakter Lösung $y(t) = e^{-t}$ numerisch und vergleiche die Verfahren.

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import matplotlib.pyplot as plt
from utils.dgl_helper import euler_explizit, runge_kutta_4, plot_vergleich
```

```python
# Rechte Seite der DGL
f = lambda t, y: -y
exakt = lambda t: np.exp(-t)

# Parameter
t0, y0 = 0.0, 1.0
h = 0.2
n = 25

t_e, y_e = euler_explizit(f, t0, y0, h, n)
t_rk, y_rk = runge_kutta_4(f, t0, y0, h, n)
```

```python
# Plot
t_fein = np.linspace(0, n * h, 500)
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(t_e, y_e, 'o--', markersize=4, label='Euler (explizit)')
ax.plot(t_rk, y_rk, 's--', markersize=4, label='Runge-Kutta 4')
ax.plot(t_fein, exakt(t_fein), 'k-', label='Exakt: $e^{-t}$')
ax.set_xlabel('t'); ax.set_ylabel('y(t)')
ax.set_title("y' = -y,  h = 0.2")
ax.legend(); ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Globaler Fehler am Endpunkt
t_end = n * h
print(f"Endpunkt t = {t_end}")
print(f"Exakte Lösung:     {exakt(t_end):.8f}")
print(f"Euler-Fehler:      {abs(y_e[-1] - exakt(t_end)):.2e}")
print(f"RK4-Fehler:        {abs(y_rk[-1] - exakt(t_end)):.2e}")
```
