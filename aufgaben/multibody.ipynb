## Hilfsfunktionen

```python
def calc_acceleration(m, x, a=None):
    """Berechne Beschleunigung fuer Planetengeschwindigkeit bei gegebener Position und Masse.

    Parameter
    ---------
    m : ndarray (N,)     – Massen
    x : ndarray (N, 3)   – Positionen
    a : ndarray (N, 3)   – Ausgabe-Array (wird bei None neu erstellt)
    """
    N = len(m)
    if a is None:
        a = np.zeros([N, 3])
    for i in range(N):
        for j in range(N):
            if j == i:
                continue
            diff = x[i, :] - x[j, :]
            diffnorm = np.linalg.norm(diff)
            a[i, :] -= G * m[j] / diffnorm**3 * diff
    return a
```

```python
def plot(planets, positions):
    """3-D-Plot der Planetentrajektorien."""
    colors = ('b', 'g', 'r', 'c', 'm', 'y', 'k')
    mpl.rcParams['legend.fontsize'] = 10
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    N = len(planets)
    for i in range(N):
        ax.plot(positions[0][:, i], positions[1][:, i], positions[2][:, i],
                '-o' + colors[i % len(colors)], label=planets[i]["name"])
    ax.legend()
    plt.tight_layout()
    plt.show()
```

```python
def ODE_solve(poshistory, m, x, v, step_method, tau, tend):
    """Einschrittverfahren von t=0 bis tend.

    Parameter
    ---------
    poshistory : list of 3 ndarrays  – wird in-place befüllt
    m          : ndarray (N,)        – Massen
    x          : ndarray (N, 3)      – Startpositionen
    v          : ndarray (N, 3)      – Startgeschwindigkeiten
    step_method: callable            – ein Schritt des Verfahrens
    tau        : float               – Zeitschrittweite
    tend       : float               – Endzeitpunkt
    """
    N = len(m)
    t = 0.0
    while t < tend - tau / 2:
        step_method(m, x, v, tau)
        t += tau
        for d in range(3):
            poshistory[d] = np.vstack([poshistory[d], x[:, d]])
        print(f"\rt = {t:.2f}", end="")
    print()
```

```python
def run_simulation(planets, method, tau, tend):
    """Aufsetzen der Startwerte, ODE-Lösung und Visualisierung."""
    m  = np.array([p["mass"]     for p in planets])
    x0 = np.array([p["position"] for p in planets])
    v0 = np.array([p["velocity"] for p in planets])

    poshistory = [None, None, None]
    for d in range(3):
        poshistory[d] = x0[:, d].reshape(1, len(planets))

    ODE_solve(poshistory, m, x0.copy(), v0.copy(), method, tau, tend)
    plot(planets, poshistory)
```
